```python
import matplotlib.pyplot as plt
import pandas as pd
import math
from IPython.display import display

df = pd.read_csv('exp_slutbetyg_kommun_2017_18.csv',
                 header=None,
                 sep=";", 
                 skiprows=5, 
                 skipfooter=33)

# display(df.head())


def reshape(df):
    # First set the index
    df.set_index([0, 1], inplace=True)
    df.index.names = ["Kommunnamn", "Typ av huvudman"]
    # Secondly parse column names as multiindex and drop the rows
    idx = pd.MultiIndex.from_tuples(list(zip(df.iloc[0], df.iloc[1])))
    df.drop(df.index[[0,1]], inplace=True)
    # TODO: Remove last column
    df.columns = idx
    return df

def clean(df):
    df.replace("..", None, inplace=True)
    df.replace(".", None, inplace=True)
    df.replace("~100", "100", inplace=True)
    # TODO: Convert comma to dot
    # TODO: Remove spaces

#     df[("Samtliga elever","Antal elever")].astype("float64")
    return df


clean(reshape(df))
```
